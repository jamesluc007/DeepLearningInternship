```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
```

```python
#cargamos librerias:

boston = load_boston()

print(boston.DESCR)
```

```python
X = np.array(boston.data[:, 5]) #Obtengo los datos de la columna 5, son las habitaciones promedio
Y = np.array(boston.target) #Obtengo los datos de precio, es el precio promedio

plt.scatter(X,Y, alpha=0.3) #Grafico en el eje X el numero de hab y en el eje Y el precio promedio
# El alpha=0.3 es para que los puntitos se vean difuminados

X = np.array([np.ones(X.size), X]).T #Con esto agrego una ilera de 1s a la matriz X
#La idea de eso es que despues funcione bien la formula de regresion lineal
#La transpuesta la agregue para que cuadren bien la matrices en la ecuacion abajo


B = np.linalg.inv(X.T @ X)@ X.T @ Y #Aca se calcula la regresion lineal
#El arroba es la multiplicacion matricial
#Es simplemente una formula matricial :D


plt.plot([4, 9], [B[0] + B[1]*4,B[0] +B[1]*9], c="red")
#Aca se grafica la curva de regresion.
#plt.plot basicamente tiene dos parametros.
#El primero es el rango del eje X, o sea va de 4 a 9
#El segundo es el rango del eje Y :D

plt.show() #Esto debe ir cada vez que se desee dibujar, por eso
#se suele poner una sola vez al final de cada cuadro

```
